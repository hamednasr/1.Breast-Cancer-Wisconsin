# Deep Learning Model to detect potato diseases
### according to 1000 images of "Early blight" and 1000 images of "late blight" and 152 images of "healthy" potatoes, a deep learning model is created and trained then deployed.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

# Cearet image dataset:

```python
potato_dataset = tf.keras.preprocessing.image_dataset_from_directory('/content/drive/MyDrive/Colab Notebooks/Computer Vision/potato disease/train',
                                                    labels='inferred',
                                                    label_mode='int',
                                                    batch_size=32,
                                                    image_size=(256, 256),
                                                    shuffle=True,
                                                    validation_split=0.1,
                                                    subset="training",
                                                    seed=55)
```

```python
potato_dataset
```

```python
classes = potato_dataset.class_names
classes
```

```python
for images, labels in potato_dataset.take(1):
  print(images.shape)
  print(labels)
```

# Visualizing images:

```python
fig , axs = plt.subplots(3,3,figsize=(10,10))

for  i, ax in enumerate(axs.flatten()):
  ax.imshow(images[i]/255)
  itemname = classes[labels[i]]
  ax.text(50, -3, itemname, fontsize=16)
  ax.set_xticks([])
  ax.set_yticks([]);
```
